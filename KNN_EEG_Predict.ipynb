```
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(EEG_df,EEG_Labels, test_size=0.3,random_state=42)

for df in [X_train,X_test,Y_test,Y_train]:
  df.reset_index(inplace=True,drop=True)

X_train_original=X_train.copy(deep=True)
X_test_original=X_test.copy(deep=True)
len(X_train_original)
```

```
from scipy.spatial.distance import cityblock


def Distance(Row1,Row2,Weights):
  Dist=0

  for i in range(len(Row1)):
    Dist+=(Weights[i]*cityblock(Row1[i],Row2[i]))
  
  return Dist
```
